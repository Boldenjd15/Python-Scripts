These are _categorical_ variables stored as numeric. Let's change the data types of these variables to `object`. Then we will use `get_dummies()` to create binary flags.

```python
cat_cols = ['SEX', 'EDUCATION', 'MARRIAGE']

for cat_col in cat_cols:
    df[cat_col] = df[cat_col].astype('str')
    
df[cat_cols].dtypes
```

Create one binary flag for each level of the categorical variable.

```python
df = pd.get_dummies(df, prefix=['SEX', 'EDU', 'MARRIAGE'])

df.head()
```

### Feature Engineering

Description of the repayment status variables:

The `PAY_` columns represent history of past payment. We tracked the past monthly payment records (from April to September, 2005) as follows: 

`PAY_0` = the repayment status in September, 2005; `PAY_2` = the repayment status in August, 2005; . . .; `PAY_6` = the repayment status in April, 2005.

The measurement scale for the repayment status is: -1 = pay duly; 1 = payment delay for one month; 2 = payment delay for two months; . . .; 8 = payment delay for eight months; 9 = payment delay for nine months and above.

```python
pay_cols = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']

for pay_col in pay_cols[:2]:
    print (df[pay_col].value_counts(sort=False))
```

Let's count how many times a customer has been in each pay category.

```python
def pay_bucket_counts(row, val):
    return (1 if row['PAY_0'] == val else 0) + \
           (1 if row['PAY_2'] == val else 0) + \
           (1 if row['PAY_3'] == val else 0) + \
           (1 if row['PAY_4'] == val else 0) + \
           (1 if row['PAY_5'] == val else 0) + \
           (1 if row['PAY_6'] == val else 0)

df['pay_bucket_0_count'] = df.apply(pay_bucket_counts, axis=1, args=(0, ))

df.head()
```

```python
# Rest of the repayment categories

df['pay_bucket_1_count'] = df.apply(pay_bucket_counts, axis=1, args=(1, ))
df['pay_bucket_2_count'] = df.apply(pay_bucket_counts, axis=1, args=(2, ))
df['pay_bucket_3_count'] = df.apply(pay_bucket_counts, axis=1, args=(3, ))
df['pay_bucket_4_count'] = df.apply(pay_bucket_counts, axis=1, args=(4, ))
df['pay_bucket_5_count'] = df.apply(pay_bucket_counts, axis=1, args=(5, ))
df['pay_bucket_6_count'] = df.apply(pay_bucket_counts, axis=1, args=(6, ))
df['pay_bucket_7_count'] = df.apply(pay_bucket_counts, axis=1, args=(7, ))
df['pay_bucket_8_count'] = df.apply(pay_bucket_counts, axis=1, args=(8, ))
```

Instead of running the above cell, we can write a function to do this.

```python
# note that there are two negative values/categories as well

for i in range(11):
    print (i-2)   
```

```python
for i in range(11):
    print (f'Processing {i-2}')
    df[f'pay_bucket_{i-2}_count'] = df.apply(pay_bucket_counts, axis=1, args=(i-2, ))
    
df.head()
```

Let's also create binary flags to indicate whether a customer ever fell into each one of the pay buckets.

```python
df['pay_bucket_0_flag'] = [1 if x > 0 else 0 for x in df['pay_bucket_0_count']]

df[['pay_bucket_0_count', 'pay_bucket_0_flag']].head()
```

```python
df['pay_bucket_1_flag'] = [1 if x > 0 else 0 for x in df['pay_bucket_1_count']]
df['pay_bucket_2_flag'] = [1 if x > 0 else 0 for x in df['pay_bucket_2_count']]
df['pay_bucket_3_flag'] = [1 if x > 0 else 0 for x in df['pay_bucket_3_count']]
df['pay_bucket_4_flag'] = [1 if x > 0 else 0 for x in df['pay_bucket_4_count']]
df['pay_bucket_5_flag'] = [1 if x > 0 else 0 for x in df['pay_bucket_5_count']]
df['pay_bucket_6_flag'] = [1 if x > 0 else 0 for x in df['pay_bucket_6_count']]
df['pay_bucket_7_flag'] = [1 if x > 0 else 0 for x in df['pay_bucket_7_count']]
df['pay_bucket_8_flag'] = [1 if x > 0 else 0 for x in df['pay_bucket_8_count']]
```

Instead of running the above cell, we can write a `for` loop to do this.

```python
# grab all columns that contains the string '_count'
pay_bucket_counts_cols = [col for col in df.columns if '_count' in col]

for pay_bucket_counts_col in pay_bucket_counts_cols:
    # create the new column name
    pay_bucket_flag = pay_bucket_counts_col.replace('count', 'flag')
    # create the new column in the dataframe
    df[pay_bucket_flag] = [1 if x > 0 else 0 for x in df[pay_bucket_counts_col]]
```

Let's create the ratio of total pay amount and bill amount.

```python
# total pay amount across all months
df['pay_amt_total'] = df[['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']].sum(axis=1)

# total bill amount across all six months
bill_amt_cols = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']

df['bill_amt_total'] = df[bill_amt_cols].sum(axis=1)

# ratio
df['pay_to_billed_ratio'] = df['pay_amt_total'] / [1 if x == 0 else x for x in df['bill_amt_total']]

df[['pay_to_billed_ratio', 'pay_amt_total', 'bill_amt_total']].head(20)
```

Create a new column that is set to 1 if the total bill amount is zero, otherwise set it to 0.

```python
df['zero_bill_amount'] = [1 if x == 0 else 0 for x in df['bill_amt_total']]

df['zero_bill_amount'].value_counts()
```

Let's create the ratio of the average billed amount to the total balance.

```python
df['utilization_avg'] = (df['bill_amt_total'] / 6) / df['LIMIT_BAL']

df[['LIMIT_BAL', 'bill_amt_total', 'utilization_avg']].head()
```

Create the ratio of the maximum billed amount to the total balance.

```python
df['utilization_max'] = df[bill_amt_cols].max(axis=1) / df['LIMIT_BAL']

df[['LIMIT_BAL', 'bill_amt_total', 'utilization_avg', 'utilization_max']].head()
```

```python
# let's drop some columns that we don't need any more

df = df.drop(pay_cols, axis=1)
```

```python
# descriptive stats

df.describe().T
```

### Tri-fold partitioning

```python
df.shape
```

Using `numpy`.

```python
import numpy as np

np.random.seed(314)

mask1 = np.random.rand(len(df)) <= .5

mask1[:10]
```

```python
# training partition: the top third
train = df[mask1]

# test and validation partition: the bottom two-third: 
test_and_valid = df[~mask1]

len(train), len(test_and_valid)
```

```python
mask2 = np.random.rand(len(test_and_valid)) <= .5

# test: the top half of test_and_valid
test = test_and_valid[mask2]

# valid: the bottom half of test_and_valid
valid = test_and_valid[~mask2]

del test_and_valid
```

```python
len(train), len(test), len(valid)
```

```python
# set-up X and y

y_train = train[target]
X_train = train.drop([target, unique_ID], axis=1)

y_test = test[target]
X_test = test.drop([target, unique_ID], axis=1)

y_valid = valid[target]
X_valid = valid.drop([target, unique_ID], axis=1)
```

Or, we can use `sklearn`.

```python
y = df[target]
X = df.drop([target, unique_ID], axis=1)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5, random_state=314)

X_test, X_valid, y_test, y_valid = train_test_split(X_test, y_test, test_size=.5, random_state=314)

len(X_train), len(X_test), len(X_valid)
```

```python
# delete the dataframe `df` as we don't need it any more

del df
```

### Initial Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier

# define classifiers
logit = LogisticRegression(solver='lbfgs', max_iter=1000, random_state=314)
tree = DecisionTreeClassifier(max_depth=4, random_state=314)
forest = RandomForestClassifier(n_estimators=100, max_depth=3, random_state=314)
gboost = GradientBoostingClassifier(random_state=314)
```

Fit the models.

```python
logit.fit(X_train, y_train)
tree.fit(X_train, y_train)
forest.fit(X_train, y_train)
gboost.fit(X_train, y_train)
```

Calculate probabilites, e.g., for the logistic regression model.

```python
logit.predict_proba(X_train)[:10]
```

Save the predicted probabilities (scores) for the train and test paritions for each model.

```python
logit_scores_train = logit.predict_proba(X_train)[:, 1]
logit_scores_test = logit.predict_proba(X_test)[:, 1]

tree_scores_train = tree.predict_proba(X_train)[:, 1]
tree_scores_test = tree.predict_proba(X_test)[:, 1]

forest_scores_train = forest.predict_proba(X_train)[:, 1]
forest_scores_test = forest.predict_proba(X_test)[:, 1]

gboost_scores_train = gboost.predict_proba(X_train)[:, 1]
gboost_scores_test = gboost.predict_proba(X_test)[:, 1]
```

### ROC Curves

```python
from sklearn.metrics import roc_curve, roc_auc_score

# calculate the false positive and true positive rates

logit_fpr_train, logit_tpr_train, _ = roc_curve(y_train, logit_scores_train)
logit_fpr_test, logit_tpr_test, _ = roc_curve(y_test, logit_scores_test)

tree_fpr_train, tree_tpr_train, _ = roc_curve(y_train, tree_scores_train)
tree_fpr_test, tree_tpr_test, _ = roc_curve(y_test, tree_scores_test)

forest_fpr_train, forest_tpr_train, _ = roc_curve(y_train, forest_scores_train)
forest_fpr_test, forest_tpr_test, _ = roc_curve(y_test, forest_scores_test)

gboost_fpr_train, gboost_tpr_train, _ = roc_curve(y_train, gboost_scores_train)
gboost_fpr_test, gboost_tpr_test, _ = roc_curve(y_test, gboost_scores_test)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='darkgrid')
plt.figure().set_size_inches(12, 9)

# train
plt.plot(logit_fpr_train, logit_tpr_train, color='royalblue', lw=2, alpha=0.4,
         label=f'Train (AUC = {roc_auc_score(y_train, logit_scores_train):0.3f})')

# test
plt.plot(logit_fpr_test, logit_tpr_test, color='royalblue', lw=2, 
         label=f'Test (AUC = {roc_auc_score(y_test, logit_scores_test):0.3f})')


plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.title('Default Risk Model: Logistic Regression', fontsize=16)
plt.legend(loc="lower right", fontsize=14)
plt.show();
```

```python
plt.figure().set_size_inches(12, 9)

plt.plot(logit_fpr_train, logit_tpr_train, color='royalblue', lw=2, alpha=0.3,
         label=f'LR Train (AUC = {roc_auc_score(y_train, logit_scores_train):0.3f})')

plt.plot(tree_fpr_train, tree_tpr_train, color='green', lw=2, alpha=0.4,
         label=f'DT Train (AUC = {roc_auc_score(y_train, tree_scores_train):0.3f})')

plt.plot(forest_fpr_train, forest_tpr_train, color='tomato', lw=2, alpha=0.4,
         label=f'RF Train (AUC = {roc_auc_score(y_train, forest_scores_train):0.3f})')

plt.plot(gboost_fpr_train, gboost_tpr_train, color='purple', lw=2, alpha=0.2,
         label=f'GB Train (AUC = {roc_auc_score(y_train, gboost_scores_train):0.3f})')

plt.plot([0, 1], [0, 1], color='gray', lw=1)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate', fontsize = 14)
plt.ylabel('True Positive Rate', fontsize = 14)
plt.title('Default Risk Model: Trainig Accuracy', fontsize = 16)
plt.legend(loc="lower right", fontsize = 14)
plt.show();
```
